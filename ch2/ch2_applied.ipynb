# 10a. Sklearn Data to Seaborn

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
print(boston.keys())
print(boston.data.shape)
print(boston.feature_names)
print(boston.DESCR)
```

```python
bos = pd.DataFrame(boston.data)
bos.columns = boston.feature_names
bos.describe()
```

# 10b.

```python
sns.pairplot(bos)
```

1. ZN is only higher than 0 when CRIM is 0. Big lots in areas where there are no crime.
2. INDUS is only consistent in areas where CRIM is above 0

# 10c. Correlations

1. CRIM appears to have strong direct correlations with ZN, INDUS, CHAS, RAD, TAX, PTRATIO
2. CRIM appears to have some sort of correlations with NOX, RM, AGE, DIS
3. CRIM seems to be not correlated with B, LSTAT
