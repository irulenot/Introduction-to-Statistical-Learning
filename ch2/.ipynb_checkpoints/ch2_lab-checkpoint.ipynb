# Contour Example

```python
xlist = np.linspace(-3.0, 3.0, 3)
ylist = np.linspace(-3.0, 3.0, 4)
print("Xlist: {}".format(xlist))

X, Y = np.meshgrid(xlist, ylist)
print("X: {}".format(X))

Z = np.sqrt(X**2 + Y**2)
cp = plt.contour(X, Y, Z)
```
