```python
# Example of a K-class where k = 2

k = 2
tk = [0, 0, 1, 0]           # all zero except one in kth place
yhat = [0.1, 0.1, 0.7, 0.1] # sum to 1
```

```python
# Calculating norm
from numpy import linalg as LA

v = yhat + tk
print(LA.norm(v))
```

The result is that we can see by classifying to the largest element of yhat we get a classification closest to our norm result. 0.7 is the closest number to 1.23 ...
